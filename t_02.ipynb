```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import matplotlib.pyplot as plt
data=pd.read_csv("Mall_Customers.csv")
df=pd.DataFrame(data)
df.info()
#age=df["Age"]
#income=df["Annual Income (k$)"]
stand=StandardScaler()
age=stand.fit_transform(df[["Age"]])
#income=stand.fit_transform(income)
age
annualinc=stand.fit_transform(df[["Annual Income (k$)"]])
df.head()
df["Age"]=age
df["Annual Income (k$)"]=annualinc
df
```

```python
KM=KMeans(n_clusters=3,random_state=42,n_init=10)
x=df[["Age","Annual Income (k$)"]]
KM.fit(x)
a=KM.labels_
b=KM.cluster_centers_
b
plt.scatter(b[:,0],b[:,1])
```

```python
X = df[["Age", "Annual Income (k$)"]]

# Plot customers, colored by their cluster
plt.scatter(X["Age"], X["Annual Income (k$)"], 
            c=KM.labels_, cmap="viridis", alpha=0.6)
```
